```python
import os
import sys
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

## load burned and tilled TS dataframes

```python
data_dir = './data/Walker2023RS/TSpoly_dataframes'
```

```python
burnTS_swir_S = pd.read_csv(os.path.join(data_dir,'TS_polyData_SentinelV4_SWIR1.csv'), index_col=0)
burnTS_swir_S['ALL'] = burnTS_swir_S.mean(axis=1)
burnTS_swir_S['stdv'] = burnTS_swir_S.std(axis=1)
burnTS_swir_S = burnTS_swir_S[burnTS_swir_S['ALL'] != 0]

tilledTS_swir_S = pd.read_csv(os.path.join(data_dir,'TSTilled_polyData_SentinelV4_SWIR1.csv'), index_col=0)
tilledTS_swir_S['ALL'] = tilledTS_swir_S.mean(axis=1)
tilledTS_swir_S['stdv'] = tilledTS_swir_S.std(axis=1)
tilledTS_swir_S = tilledTS_swir_S[tilledTS_swir_S['ALL'] != 0]
```

```python
fig = plt.figure(figsize=(7, 5))
plt.errorbar(burnTS_swir_S.index, burnTS_swir_S['ALL'], yerr=burnTS_swir_S['stdv'], fmt='o', color='k', label='burned fields')
plt.errorbar(tilledTS_swir_S.index, tilledTS_swir_S['ALL'], yerr=tilledTS_swir_S['stdv'], fmt='o', color='g', markerfacecolor='none', label='unburned fields')
plt.title ("b) Sentinel-2 SWIR1", fontsize=20, fontname = 'Palatino Linotype', loc='left')
plt.ylabel ('SWIR1', fontsize=18, fontname='Palatino Linotype')
plt.xlabel ('Days from event (burn or till))', fontsize=18, fontname='Palatino Linotype')
plt.xlim(-40,40)
plt.legend(loc='upper right', fontsize=14, frameon=False, handletextpad=0.1);
```

```python
burnTS_nir_S = pd.read_csv(os.path.join(data_dir,'TS_polyData_SentinelV4_nir.csv'), index_col=0)
burnTS_nir_S['ALL'] = burnTS_nir_S.mean(axis=1)
burnTS_nir_S['stdv'] = burnTS_nir_S.std(axis=1)
burnTS_nir_S = burnTS_nir_S[burnTS_nir_S['ALL'] != 0]

tilledTS_nir_S = pd.read_csv(os.path.join(data_dir,'TSTilled_polyData_SentinelV4_nir.csv'), index_col=0)
tilledTS_nir_S['ALL'] = tilledTS_nir_S.mean(axis=1)
tilledTS_nir_S['stdv'] = tilledTS_nir_S.std(axis=1)
tilledTS_nir_S = tilledTS_nir_S[tilledTS_nir_S['ALL'] != 0]
```

```python
fig = plt.figure(figsize=(7, 5))
plt.errorbar(burnTS_nir_S.index, burnTS_nir_S['ALL'], yerr=burnTS_nir_S['stdv'], fmt='o', color='k', label='burned fields')
plt.errorbar(tilledTS_nir_S.index, tilledTS_nir_S['ALL'], yerr=tilledTS_nir_S['stdv'], fmt='o', color='g', markerfacecolor='none', label='unburned fields')
plt.title ("Sentinel-2 nir values for burned and unburned rice paddys across time", fontsize=12)
plt.ylabel ('nir (band8)', fontsize=10)
plt.xlabel ('Days from event (burn or till))', fontsize=10)
plt.xlim(-40,40)
plt.legend(loc='upper right');
```

```python
burnTS_nir_P = pd.read_csv(os.path.join(data_dir,'TS_polyData_PlanetV4_nir.csv'), index_col=0)
burnTS_nir_P['ALL'] = burnTS_nir_P.mean(axis=1)
burnTS_nir_P['stdv'] = burnTS_nir_P.std(axis=1)
burnTS_nir_P = burnTS_nir_P[burnTS_nir_P['ALL'] != 0]

tilledTS_nir_P = pd.read_csv(os.path.join(data_dir,'TSTilled_polyData_PlanetV4_nir.csv'), index_col=0)
tilledTS_nir_P['ALL'] = tilledTS_nir_P.mean(axis=1)
tilledTS_nir_P['stdv'] = tilledTS_nir_P.std(axis=1)
tilledTS_nir_P = tilledTS_nir_P[tilledTS_nir_P['ALL'] != 0]
```

```python
fig = plt.figure(figsize=(7, 5))
plt.errorbar(burnTS_nir_P.index, burnTS_nir_P['ALL'], yerr=burnTS_nir_P['stdv'], fmt='o', color='k', label='burned fields')
plt.errorbar(tilledTS_nir_P.index, tilledTS_nir_P['ALL'], yerr=tilledTS_nir_P['stdv'], fmt='o', color='g',markerfacecolor='none', label='unburned fields')
plt.title ("PlanetScope nir values for burned and unburned rice paddys across time", fontsize=12)
plt.ylabel ('nir (band8)', fontsize=10)
plt.xlabel ('Days from event (burn or till))', fontsize=10)
plt.xlim(-40,40)
plt.legend(loc='upper right');
```
