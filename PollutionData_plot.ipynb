## To summarize pollution data for time period
After downloading pollution data via `PollutionData_access` notebook
Note: preloaded data was downloaded from `https://dataspace.princeton.edu/handle/88435/dsp01zw12z843p` because india pollution dashboard website was not working

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pollutant = 'PM2.5' #options = 'PM2.5' | 'PM10' |'
pollution_data = pd.read_csv('./data/PM25/Punjab_Bhatinda.csv')
pollution_data['PM2.5'] = pd.to_numeric(pollution_data['PM2.5'], errors='coerce')
pollution_data['From Date'] = pd.to_datetime(pollution_data['From Date'])
pollution_ts = pollution_data[pollution_data['From Date'] > pd.Timestamp(2019,10,1)]
#pollution_ts.set_index('From Date', inplace=True)
pollution_daily_Bhatinda = pollution_ts.set_index('From Date').resample('D')[pollutant].mean()
pollution_daily_Bhatinda.sort_values(ascending = False).head(n=20)
```
